```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Insert the average execution times ---
labels = [
    'Normalisation', 
    'Noise Adding', 
    'Percentile Clipping', 
    'Min-Max Norm', 
    'Masked Norm'
]

# average execution times for each operation (in seconds)
np_times = [1.0410, 2.2657, 2.1569, 0.5762, 0.4487]
rust_times = [0.8900, 0.3708, 1.4332, 0.3504, 0.2779]

# calculated speedup for each operation
speedups = [1.17, 6.11, 1.50, 1.64, 1.61]

# --- 2. Set up the chart and bar chart ---
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 7))

rects1 = ax.bar(x - width/2, np_times, width, label='NumPy', color='#4C72B0', edgecolor='white')
rects2 = ax.bar(x + width/2, rust_times, width, label='Rust (Fast Med Vision)', color='#DD8452', edgecolor='white')

# --- 3. Set labels and title ---
ax.set_ylabel('Average Execution Time (Seconds)', fontsize=12, fontweight='bold')
ax.set_title('Performance Comparison: NumPy vs. Rust Engine\nData: 5 Patients, Shape: (100, 512, 512) | 10 Iterations Avg', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=11, fontweight='bold')
ax.legend(fontsize=12)

# --- 4. Labelling ---
ax.bar_label(rects1, padding=3, fmt='%.4fs', fontsize=10)
ax.bar_label(rects2, padding=3, fmt='%.4fs', fontsize=10)

for i in range(len(labels)):
    max_height = max(np_times[i], rust_times[i])
    ax.text(x[i], max_height + 0.15, f'{speedups[i]}x Faster', 
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='#D62728')

ax.set_ylim(0, max(np_times) + 0.6)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# disable top and right spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

plt.show()
```
